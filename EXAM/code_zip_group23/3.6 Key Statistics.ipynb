```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Add-on toolkit for plt
```

```python
data = pd.read_csv('cleaned_data.csv', encoding='utf-16')
```

```python
data.columns
```

```python
# Select specific columns
selected_columns = ['price', 'living_space', 'rooms']
selected_data = data[selected_columns]

# Calculate summary statistics for selected columns
statistics = {
    'Metric': ['Mean', 'Standard Deviation', 'Minimum', '25% Percentile', '50% Percentile (Median)', '75% Percentile', 'Maximum']
}

for column in selected_columns:
    statistics[column] = [
        selected_data[column].mean(),
        selected_data[column].std(),
        selected_data[column].min(),
        selected_data[column].quantile(0.25),
        selected_data[column].quantile(0.50),
        selected_data[column].quantile(0.75),
        selected_data[column].max()
    ]

# Create a summary DataFrame
summary_df = pd.DataFrame(statistics)

# Round the values to 2 decimal places
summary_df = summary_df.round(2)

# Display the summary DataFrame
print(summary_df)
```

```python
# Export the summary DataFrame to a CSV file
summary_df.to_csv('summary_statistics.csv', index=False)
```
